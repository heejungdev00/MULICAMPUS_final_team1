```python
def generate_instruction(angles):
    # 각 부위의 이름
    parts = ["오른 팔꿈치", "왼쪽 팔꿈치", "오른쪽 무릎", "왼쪽 무릎", "오른쪽 발목", "왼쪽 발목", "오른쪽 골반", "왼쪽 골반"]
    
    # 각도에 따라 구부려야 하는 부위와 펴야 하는 부위를 구분
    bent_parts = []
    straight_parts = []
    
    for i, angle in enumerate(angles):
        if angle > 0:
            bent_parts.append(parts[i])
        elif angle < 0:
            straight_parts.append(parts[i])
    
    # 문구 생성
    if bent_parts:
        bent_instruction = "와 ".join(bent_parts) + "을(를) 더 구부려주시고"
    else:
        bent_instruction = ""
    
    if straight_parts:
        straight_instruction = "와 ".join(straight_parts) + "을(를) 더 펴주세요."
    else:
        straight_instruction = ""
    
    # 문구 조합
    if bent_instruction and straight_instruction:
        instruction = f"{bent_instruction} {straight_instruction}"
    elif bent_instruction:
        instruction = f"{bent_instruction[:-2]} 해주세요."  # 마지막의 "와" 제거
    elif straight_instruction:
        instruction = straight_instruction
    
    return instruction

# 예시 배열
angles = [10, -5, 20, -15, 8, -3, 12, -7]
instruction = generate_instruction(angles)
print(instruction)
```

```python
import random

# 랜덤 각도 배열 생성
def generate_random_angles(size=8, low=-45, high=45):
    return [random.randint(low, high) for _ in range(size)]

# 예시: 256개의 배열 생성 및 문구 생성
instructions = []
arr= []
for _ in range(50):
    angles = generate_random_angles()
    arr.append(angles)
    instruction = generate_instruction(angles)
    instructions.append(instruction)

# 예시: 첫 번째 문구 출력
print(instructions[0])
```
