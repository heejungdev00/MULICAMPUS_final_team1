```python
import os
import pydicom
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import matplotlib.pyplot as plt
from tqdm import tqdm
from PIL import Image
```

```python
def resize_image(image, size=(128, 128)):
    img = Image.fromarray(image).convert('L')  # Convert to grayscale
    return np.array(img.resize(size, Image.Resampling.LANCZOS))
```

```python
def load_dicom_images(base_path, series_df, batch_size=32, image_size=(128, 128)):
    images = []
    labels = []
    count = 0
    for _, row in tqdm(series_df.iterrows(), total=series_df.shape[0]):
        study_id = row['study_id']
        series_id = row['series_id']
        instance_number = row['instance_number']
        condition = row['condition']
        level = row['level']
        
        file_path = os.path.join(base_path, str(study_id), str(series_id), f'{instance_number}.dcm')
        if os.path.exists(file_path):
            ds = pydicom.dcmread(file_path)
            resized_image = resize_image(ds.pixel_array, size=image_size)
            images.append(resized_image)
            labels.append(f"{condition}_{level}")
            count += 1
        
        if count == batch_size:
            images = np.array(images) / 255.0
            images = np.expand_dims(images, axis=-1)  # Add channel dimension
            yield images, labels
            images = []
            labels = []
            count = 0

    if count > 0:
        images = np.array(images) / 255.0
        images = np.expand_dims(images, axis=-1)  # Add channel dimension
        yield images, labels
```

```python
# Define the path to the dataset
path = 'spine/'

# Read the CSV files
df_train_main = pd.read_csv(os.path.join(path, 'train.csv'))
df_train_label = pd.read_csv(os.path.join(path, 'train_label_coordinates.csv'))
df_train_desc = pd.read_csv(os.path.join(path, 'train_series_descriptions.csv'))
df_test_desc = pd.read_csv(os.path.join(path, 'test_series_descriptions.csv'))
df_sub = pd.read_csv(os.path.join(path, 'sample_submission.csv'))

# Display the first few rows of each dataframe to ensure they loaded correctly
print(df_train_main.head())
print(df_train_label.head())
print(df_train_desc.head())
print(df_test_desc.head())
print(df_sub.head())
```

```python
# Define paths and load train images in batches
base_path_train = os.path.join(path, 'train_images')
batch_size = 32
image_size = (128, 128)
image_generator = load_dicom_images(base_path_train, df_train_label, batch_size=batch_size, image_size=image_size)

images_all = []
labels_all = []

for images_batch, labels_batch in image_generator:
    images_all.append(images_batch)
    labels_all.append(labels_batch)

images = np.concatenate(images_all, axis=0)
labels = np.concatenate(labels_all, axis=0)

print(f"Loaded {images.shape[0]} images.")
```

```python
# Encode labels
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(images, encoded_labels, test_size=0.2, random_state=42)
```

```python
# Define the CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(image_size[0], image_size[1], 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(np.unique(encoded_labels)), activation='softmax')  
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=5, validation_data=(X_val, y_val), batch_size=32)
```

```python
# Plot training history
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
